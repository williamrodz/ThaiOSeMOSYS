```python
#Plot a graph from a CSV file using matplotlib
import matplotlib.pyplot as plt
import csv

#Enter the name of the csv file 
csvfile = input('Enter the name of the CSV file: ')

#Open the CSV file
with open(csvfile) as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    
    #Create empty lists to store the data
    years = []
    emissions = []
    
    # Find the column number of the "y" column
    header = next(readCSV)
    column_year = header.index("y")
    column_emission = header.index("AnnualEmissions")
    
    # Read the data from the CSV file
    for row in readCSV:
        year = row[column_year]
        emission = row[column_emission]
        years.append(year)
        emissions.append(emission)

#Plot the data
plt.plot(years, emissions, label='Annual Emissions')
plt.xlabel('Year')
plt.ylabel('Emissions')
plt.title('Annual Emissions')
plt.legend()
plt.xticks(years[::5])
plt.yticks(emissions[::10])
plt.show()
```
